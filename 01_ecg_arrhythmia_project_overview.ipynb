# Rozpoznawanie nieprawidłowych uderzeń serca na podstawie sygnału EKG

**Przedmiot:** Akwizycja danych biomedycznych  
**Autorzy:** Ewa Kędziera, Jan Rapacz

## Cel projektu

Celem projektu jest pokazanie, w jaki sposób z rzeczywistego sygnału biomedycznego można przejść od akwizycji i opisu fizjologicznego do przygotowania danych oraz klasyfikacji uderzeń serca.

Projekt dotyczy sygnału elektrokardiograficznego, czyli EKG. Analizowana jest informacja zawarta w zapisie aktywności elektrycznej serca, ze szczególnym uwzględnieniem:

- położenia uderzeń serca w czasie,
- odstępów RR opisujących rytm,
- kształtu krótkich fragmentów sygnału wokół uderzeń,
- oznaczeń eksperckich dostępnych w bazie danych.

Docelowo projekt ma prowadzić do przygotowania kompletnego pipeline’u analizy zapisu EKG, obejmującego:

1. wczytanie i wizualizację sygnału EKG,
2. ocenę jakości dostępnych danych,
3. preprocessing sygnału,
4. analizę sygnału z wykorzystaniem różnych reprezentacji danych,
5. klasyfikację uderzeń lub fragmentów sygnału jako prawidłowych albo potencjalnie nieprawidłowych,
6. ocenę skuteczności zastosowanych metod.

## Zaplecze biomedyczne

### 1. Czym jest EKG?

**EKG**, czyli **elektrokardiografia**, jest metodą rejestrowania aktywności elektrycznej serca. Każdy cykl pracy serca jest związany z przewodzeniem impulsu elektrycznego przez układ bodźcoprzewodzący i mięsień sercowy. Elektrody umieszczone na powierzchni skóry mierzą różnice potencjałów elektrycznych, a wynik jest zapisywany jako sygnał zmieniający się w czasie.

EKG jest sygnałem biomedycznym szczególnie istotnym w analizie rytmu serca, ponieważ pozwala obserwować:

- regularność kolejnych uderzeń,
- częstość pracy serca,
- zaburzenia przewodzenia,
- zmiany kształtu zespołów QRS,
- dodatkowe lub przedwczesne pobudzenia.

EKG jest badaniem:

- nieinwazyjnym,
- szybkim,
- szeroko stosowanym klinicznie,
- używanym m.in. do oceny rytmu serca, częstości pracy serca i wykrywania zaburzeń przewodzenia.

### 2. Podstawowe elementy sygnału EKG

Typowy pojedynczy cykl pracy serca w zapisie EKG składa się z kilku charakterystycznych elementów:

| Element | Znaczenie fizjologiczne | Znaczenie w analizie sygnału |
|---|---|---|
| **załamek P** | depolaryzacja przedsionków | informacja o pobudzeniu przedsionków |
| **zespół QRS** | depolaryzacja komór | najbardziej wyraźny fragment sygnału |
| **pik R** | zwykle najwyższy punkt zespołu QRS | punkt odniesienia do analizy rytmu |
| **załamek T** | repolaryzacja komór | informacja o fazie powrotu komór do stanu spoczynkowego |
| **odstęp RR** | czas między kolejnymi pikami R | podstawowa cecha rytmu serca |
| **odstęp PR/QT** | parametry przewodzenia i repolaryzacji | potencjalnie przydatne w bardziej zaawansowanej analizie |

W projekcie najważniejsze są dwa rodzaje informacji:

1. **informacja rytmiczna** - położenia kolejnych uderzeń i odstępy RR,
2. **informacja morfologiczna** - kształt fragmentu sygnału wokół uderzenia.

### 3. Akwizycja sygnału EKG

Sygnał EKG pozyskuje się za pomocą elektrod umieszczonych na powierzchni skóry. Elektrody rejestrują różnice potencjałów elektrycznych powstające podczas pracy serca.

Typowy tor akwizycji EKG obejmuje:

1. **elektrody pomiarowe**,
2. **kontakt elektroda–skóra**,
3. **wzmacniacz sygnału biologicznego**,
4. **filtrację analogową lub cyfrową**,
5. **przetwornik analogowo-cyfrowy**,
6. **zapis cyfrowy sygnału**,
7. **dalszą analizę komputerową**.

W pomiarach laboratoryjnych duże znaczenie mają:

- prawidłowe rozmieszczenie elektrod,
- przygotowanie skóry,
- ograniczenie ruchów osoby badanej,
- jakość przewodów i połączeń,
- ustawienia częstotliwości próbkowania,
- kontrola zakłóceń.

### 4. Odprowadzenia EKG

EKG może być rejestrowane w różnych konfiguracjach odprowadzeń.

Przykłady:

- **jednoodprowadzeniowe EKG** - często używane w prostych systemach monitorujących,
- **kilkuodprowadzeniowe EKG** - np. urządzenia przenośne, Holter,
- **12-odprowadzeniowe EKG** - standard kliniczny dający pełniejszy obraz aktywności elektrycznej serca.

W bazie MIT-BIH każdy rekord zawiera dwa kanały EKG. Najczęściej wykorzystywanym kanałem jest zmodyfikowane odprowadzenie II, oznaczane jako **MLII**, ponieważ zwykle dobrze eksponuje zespoły QRS i piki R.

W projekcie wybór kanału jest decyzją przetwarzania danych: dla prostszego i bardziej kontrolowanego wariantu można analizować jeden kanał, a w rozszerzeniu wykorzystać oba kanały.

### 5. Zakłócenia i artefakty w sygnale EKG

Sygnał EKG jest podatny na zakłócenia, które mogą utrudniać zarówno interpretację biomedyczną, jak i automatyczną klasyfikację.

| Rodzaj zakłócenia | Możliwe źródło | Skutek w sygnale |
|---|---|---|
| **dryft linii bazowej** | oddychanie, ruch elektrod, zmiana impedancji skóry | wolne przesuwanie sygnału w górę lub w dół |
| **zakłócenia sieciowe 50/60 Hz** | zasilanie elektryczne | regularna szum o częstotliwości sieci |
| **artefakty ruchowe** | ruch pacjenta lub przewodów | nagłe skoki i deformacje sygnału |
| **szum mięśniowy EMG** | napięcie mięśni | zakłócenia wysokoczęstotliwościowe |
| **słaby kontakt elektrody** | odklejenie lub nieprawidłowe przygotowanie skóry | niestabilny, zaszumiony zapis |

Z tego powodu w analizie EKG można zastosować np.:

- filtrację pasmowoprzepustową,
- filtr notch dla zakłóceń sieciowych,
- odrzucanie fragmentów bardzo zaszumionych,
- normalizację amplitudy.

### 6. Czym jest arytmia?

**Arytmia serca** oznacza zaburzenie rytmu lub częstości pracy serca. W prawidłowych warunkach impulsy elektryczne powstają i przewodzą się w uporządkowany sposób. W przypadku arytmii impulsy mogą pojawiać się:

- zbyt szybko,
- zbyt wolno,
- nieregularnie,
- w nieprawidłowym miejscu układu bodźcoprzewodzącego.

W efekcie rytm serca może stać się nieregularny albo odbiegać od typowego rytmu zatokowego.

W zapisie EKG arytmie mogą objawiać się na dwa główne sposoby:

1. **zmianą rytmu**, czyli nietypowymi odstępami między kolejnymi uderzeniami,
2. **zmianą morfologii**, czyli innym kształtem zespołu QRS lub fragmentu sygnału wokół uderzenia.

Ta obserwacja bezpośrednio uzasadnia podział projektu na dwie części: analizę rytmu na podstawie odstępów RR oraz analizę bogatszej reprezentacji sygnału na podstawie cech i okien czasowych.

### 7. Przykładowe zaburzenia widoczne w EKG

| Zaburzenie | Ogólny opis | Możliwa cecha w EKG |
|---|---|---|
| **bradykardia** | zbyt wolna częstość pracy serca | wydłużone odstępy RR |
| **tachykardia** | zbyt szybka częstość pracy serca | skrócone odstępy RR |
| **pobudzenia przedwczesne** | dodatkowe uderzenia pojawiające się wcześniej niż oczekiwano | skrócony RR przed uderzeniem, często pauza po uderzeniu |
| **pobudzenia komorowe** | uderzenia pochodzące z komór | często zmieniony kształt QRS |
| **pobudzenia nadkomorowe** | uderzenia pochodzące powyżej komór | często zaburzony rytm, czasem subtelna zmiana kształtu |
| **migotanie przedsionków** | nieregularna praca przedsionków | nieregularne RR, brak typowych załamków P |
| **częstoskurcz komorowy** | szybki rytm pochodzenia komorowego | szerokie zespoły QRS, szybki rytm |
| **bloki przewodzenia** | zaburzenia przewodzenia impulsu | zmieniona morfologia, zwłaszcza zespołu QRS |

Dla projektu rozsądne jest zawężenie problemu do jednej z dwóch wersji:

1. **klasyfikacja binarna:** uderzenie prawidłowe vs nieprawidłowe,
2. **klasyfikacja kilku wybranych typów uderzeń:** np. normalne, nadkomorowe, komorowe.

## Koncepcja projektu

### 1. Dane

Do realizacji projektu wybrano publiczną bazę **MIT-BIH Arrhythmia Database**. Jest to klasyczny zbiór rzeczywistych zapisów EKG, często wykorzystywany w badaniach nad automatyczną analizą arytmii.

Baza jest odpowiednia do projektu, ponieważ zawiera:

- rzeczywiste, a nie syntetyczne sygnały EKG,
- adnotacje ekspertów wskazujące położenia uderzeń,
- symbole typów uderzeń,
- dwa kanały sygnału,
- rekordy pochodzące od różnych pacjentów,
- gotową obsługę w bibliotece `wfdb`.

### 2. Idea projektu

Główna idea projektu polega na porównaniu dwóch sposobów reprezentowania informacji zawartej w sygnale EKG.

```text
sygnał EKG + adnotacje
→ przygotowanie danych
→ reprezentacja rytmu RR
→ klasyczny baseline regułowy

sygnał EKG + adnotacje
→ przygotowanie danych
→ reprezentacja fragmentów sygnału wokół uderzeń
→ klasyczne ML na cechach oraz deep learning na oknach sygnału
```

Pierwsza reprezentacja jest prosta i interpretowalna, ale opisuje głównie rytm. Druga reprezentacja jest bogatsza, ponieważ uwzględnia także morfologię uderzenia. Dzięki temu projekt pokazuje, jak zmiana reprezentacji danych wpływa na możliwości klasyfikacji.

### 3. Zakres i ograniczenia projektu

Projekt ma charakter edukacyjno-badawczy. Jego celem jest analiza sygnału biomedycznego i porównanie metod przetwarzania danych, a nie stworzenie gotowego narzędzia diagnostycznego.

Najważniejsze ograniczenia:

- wykorzystywane są adnotowane pozycje uderzeń z bazy MIT-BIH,
- klasyfikacja dotyczy pojedynczych uderzeń, a nie pełnej diagnozy pacjenta,
- baza MIT-BIH jest dobrze udokumentowana, ale stosunkowo mała,
- rozkład klas jest niezbalansowany,
- część rekordów zawiera uderzenia stymulowane rozrusznikiem,
- wyniki wymagają ostrożnej interpretacji biomedycznej.
