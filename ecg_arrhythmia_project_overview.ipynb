# Akwizycja Danych Biomedycznych - projekt

## Koncepcja projektu

*Temat:* **Rozpoznawanie wybranych zaburzeń rytmu serca na podstawie sygnału EKG z wykorzystaniem metod przetwarzania sygnałów i uczenia maszynowego**

Projekt dotyczy analizy sygnału elektrokardiograficznego, czyli EKG. Główna idea polega na tym, aby na podstawie rzeczywistego zapisu EKG przygotować pipeline obejmujący:

1. wczytanie i wizualizację sygnału,
2. analizę jakości danych,
3. preprocessing sygnału,
4. wyznaczenie położeń pików R na podstawie adnotacji lub algorytmu detekcji,
5. analizę odstępów RR jako klasyczne podejście do oceny regularności rytmu serca,
6. przygotowanie okien sygnału wokół uderzeń serca na potrzeby metod uczenia maszynowego lub deep learningu,
7. klasyfikację uderzeń lub fragmentów sygnału jako prawidłowych albo potencjalnie nieprawidłowych,
8. ocenę skuteczności zastosowanych metod.

## 1. Czym jest EKG?

**EKG**, czyli **elektrokardiografia**, jest metodą rejestrowania aktywności elektrycznej serca. Każdy cykl pracy serca jest związany z przewodzeniem impulsu elektrycznego przez układ bodźcoprzewodzący i mięsień sercowy. Elektrody umieszczone na powierzchni skóry mierzą różnice potencjałów elektrycznych, a wynik jest zapisywany jako sygnał zmieniający się w czasie.

EKG jest badaniem:

- nieinwazyjnym,
- szybkim,
- szeroko stosowanym klinicznie,
- używanym m.in. do oceny rytmu serca, częstości pracy serca i wykrywania zaburzeń przewodzenia.

## 2. Podstawowe elementy sygnału EKG

Typowy pojedynczy cykl serca w sygnale EKG składa się z kilku charakterystycznych fragmentów:

| Element | Znaczenie fizjologiczne | Znaczenie w analizie sygnału |
|---|---|---|
| **załamek P** | depolaryzacja przedsionków | może być istotny przy analizie rytmu przedsionkowego |
| **zespół QRS** | depolaryzacja komór | zwykle najbardziej wyraźny fragment sygnału |
| **pik R** | najwyższy punkt zespołu QRS w wielu odprowadzeniach | punkt odniesienia do wyznaczania rytmu |
| **załamek T** | repolaryzacja komór | może być istotny przy analizie zmian morfologii |
| **odstęp RR** | czas między kolejnymi pikami R | podstawowa cecha rytmu serca |
| **odstęp PR/QT** | parametry przewodzenia i repolaryzacji | potencjalnie przydatne w bardziej zaawansowanej analizie |

W prostym projekcie najważniejsze będą:

- piki R,
- odstępy RR,
- kształt fragmentu sygnału wokół zespołu QRS.

## 3. Czym jest arytmia serca?

**Arytmia serca** oznacza zaburzenie rytmu lub częstości pracy serca. W prawidłowych warunkach impulsy elektryczne powstają i przewodzą się w uporządkowany sposób. W przypadku arytmii impulsy mogą pojawiać się:

- zbyt szybko,
- zbyt wolno,
- nieregularnie,
- w nieprawidłowym miejscu układu bodźcoprzewodzącego.

W efekcie rytm serca może stać się nieregularny albo odbiegać od typowego rytmu zatokowego.

## 4. Przykładowe zaburzenia rytmu serca

| Zaburzenie | Ogólny opis | Możliwe cechy w EKG |
|---|---|---|
| **bradykardia** | zbyt wolny rytm serca | wydłużone odstępy RR |
| **tachykardia** | zbyt szybki rytm serca | skrócone odstępy RR |
| **ekstrasystolie / pobudzenia przedwczesne** | dodatkowe uderzenia pojawiające się wcześniej niż oczekiwano | nieregularność RR, zmieniony kształt QRS |
| **migotanie przedsionków** | nieregularna praca przedsionków | nieregularne RR, brak typowych załamków P |
| **częstoskurcz komorowy** | szybki rytm pochodzenia komorowego | szerokie zespoły QRS, szybki rytm |
| **bloki przewodzenia** | zaburzenia przewodzenia impulsów | zmiany odstępów i morfologii zespołów |

Dla projektu rozsądne jest zawężenie problemu do jednej z dwóch wersji:

1. **klasyfikacja binarna:** uderzenie prawidłowe vs nieprawidłowe,
2. **klasyfikacja kilku wybranych typów uderzeń:** np. normalne, nadkomorowe, komorowe.

## 5. Jak pozyskuje się sygnał EKG?

Sygnał EKG pozyskuje się za pomocą elektrod umieszczonych na powierzchni skóry. Elektrody rejestrują różnice potencjałów elektrycznych powstające podczas pracy serca.

Typowy tor akwizycji EKG obejmuje:

1. **elektrody pomiarowe**,
2. **kontakt elektroda–skóra**,
3. **wzmacniacz sygnału biologicznego**,
4. **filtrację analogową lub cyfrową**,
5. **przetwornik analogowo-cyfrowy**,
6. **zapis cyfrowy sygnału**,
7. **dalszą analizę komputerową**.

W pomiarach laboratoryjnych duże znaczenie mają:

- prawidłowe rozmieszczenie elektrod,
- przygotowanie skóry,
- ograniczenie ruchów osoby badanej,
- jakość przewodów i połączeń,
- ustawienia częstotliwości próbkowania,
- kontrola zakłóceń.

## 6. Odprowadzenia EKG

EKG może być rejestrowane w różnych konfiguracjach odprowadzeń.

Przykłady:

- **jednoodprowadzeniowe EKG** - często używane w prostych systemach monitorujących,
- **kilkuodprowadzeniowe EKG** - np. urządzenia przenośne, Holter,
- **12-odprowadzeniowe EKG** - standard kliniczny dający pełniejszy obraz aktywności elektrycznej serca.

## 7. Zakłócenia w sygnale EKG

Sygnał EKG jest podatny na wiele zakłóceń. To bardzo ważny element projektu, ponieważ pokazuje praktyczny aspekt akwizycji danych biomedycznych.

| Rodzaj zakłócenia | Możliwe źródło | Skutek w sygnale |
|---|---|---|
| **dryft linii bazowej** | oddychanie, ruch elektrod, zmiana impedancji skóry | wolne przesuwanie się sygnału w górę/dół |
| **zakłócenia sieciowe 50/60 Hz** | zasilanie elektryczne | regularny szum o częstotliwości sieci |
| **artefakty ruchowe** | ruch pacjenta, poruszanie przewodami | nagłe skoki i deformacje sygnału |
| **szum mięśniowy EMG** | napięcie mięśni | szum wysokoczęstotliwościowy |
| **słaby kontakt elektrody** | nieprawidłowe przyklejenie elektrod | niestabilny, zaszumiony sygnał |

W preprocessingu można zastosować np.:

- filtrację pasmowoprzepustową,
- filtr notch dla zakłóceń sieciowych,
- odrzucanie fragmentów bardzo zaszumionych,
- normalizację amplitudy.

## 8. Dane do projektu

Dobrym wyborem jest publiczna baza:

### MIT-BIH Arrhythmia Database

Baza MIT-BIH Arrhythmia Database zawiera rzeczywiste zapisy EKG z adnotacjami uderzeń serca. Jest często wykorzystywana w projektach i badaniach dotyczących detekcji arytmii.

Najważniejsze cechy tej bazy:

- zapisy EKG są rzeczywiste, a nie syntetyczne,
- dane zawierają adnotacje położenia uderzeń,
- dostępne są symbole klas uderzeń,
- baza jest obsługiwana przez bibliotekę `wfdb` w Pythonie.

W projekcie można użyć bazy MIT-BIH jako źródła danych zamiast wykonywania własnych pomiarów.

## 9. Ogólny plan projektu

Projekt można zrealizować etapami.

### Etap 1 - analiza literatury i opis biomedyczny

- czym jest EKG,
- jak powstaje sygnał EKG,
- jakie są podstawowe elementy sygnału,
- czym są arytmie,
- jakie zaburzenia można obserwować w EKG.

### Etap 2 - pozyskanie danych

- wybór bazy danych, np. MIT-BIH Arrhythmia Database,
- wczytanie sygnałów w Pythonie,
- wczytanie adnotacji,
- wybór jednego lub kilku rekordów do analizy początkowej.

### Etap 3 - EDA

- wizualizacja surowego sygnału,
- sprawdzenie częstotliwości próbkowania,
- sprawdzenie dostępnych kanałów,
- analiza rozkładu klas uderzeń,
- analiza liczby uderzeń w rekordach,
- wizualizacja przykładów uderzeń prawidłowych i nieprawidłowych.

### Etap 4 - preprocessing

- wybór kanału EKG,
- filtracja pasmowoprzepustowa, np. 0.5–40 Hz,
- opcjonalnie filtr notch dla 50/60 Hz,
- użycie adnotacji lub detekcja pików R,
- obliczenie odstępów RR,
- segmentacja uderzeń wokół pików R,
- normalizacja sygnału.

### Etap 5 - podejście klasyczne

- analiza odstępów RR,
- obliczenie chwilowego rytmu serca,
- wykrywanie nieregularności prostymi regułami,
- porównanie z adnotacjami.

### Etap 6 - uczenie maszynowe

- ekstrakcja cech z pojedynczych uderzeń,
- przygotowanie etykiet,
- podział train/test na poziomie rekordów,
- trenowanie klasyfikatorów, np. Random Forest, SVM, Logistic Regression,
- ocena wyników.

### Etap 7 - deep learning (opcjonalnie)

- przygotowanie okien sygnału jako wejścia do modelu,
- budowa prostej sieci 1D CNN,
- porównanie wyników z metodami klasycznymi.

### Etap 8 - podsumowanie i wnioski

- porównanie metod,
- omówienie ograniczeń,
- wskazanie możliwych rozszerzeń.
